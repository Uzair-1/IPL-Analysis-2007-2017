```python
def batmanRun_againstTeams(batman_name):
    player = delivery[delivery['batsman']==batman_name]
    return player.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending =  False).head(5)
```

```python
# Find the most destructive death-over(16-20) batsman in the history of IPL
# Find the strike rate and also have a condition the batsman play 200 balls
```

```python
delivery.head(1)
```

```python
death_over= delivery[delivery['over'] > 15]
death_over
```

```python
player_play_balls = death_over.groupby('batsman')['batsman_runs'].count()
x = player_play_balls > 200
batsman_list=player_play_balls[x].index.tolist()
batsman_list
```

```python
final_df=delivery[delivery['batsman'].isin(batsman_list)]
final_df.shape
```

```python
runs = final_df.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final_df.groupby('batsman')['batsman_runs'].count()
```

```python
final_result = (runs/balls)*100
```

```python
final_result.sort_values(ascending =  False)
```
